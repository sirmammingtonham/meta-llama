```python
import pandas as pd
import os

metadata = pd.read_json("../bigbench_metadata/metadata.json")
metadata.head() # lots of columns because includes subtasks
```

```python
train_tasks = [f.rsplit('.')[0] for f in os.listdir("../training_data/")]
```

```python
diff = len(set(train_tasks).difference(set(metadata.columns)))
print(f'set difference: {diff}')
```

```python
out = {t: [metadata[t]['description']] for t in train_tasks}
```

```python
df = pd.DataFrame(out)
# df = df.T 
df = df.T
df = df.reset_index()
df.columns = ["task", "description"]
```

```python
df.to_csv("task_descriptions.csv")
```
